## Limpieza y preparación de datos

```python
import pandas as pd
import numpy as np
import os
```

```python
# Definir rutas

raw_path = "/workspaces/tumipay_analysis/data/raw"
processed_path = "/workspaces/tumipay_analysis/data/processed"
```

```python
# Cargar datasets

clientes = pd.read_csv(f"{raw_path}/clientes.csv")
creditos = pd.read_csv(f"{raw_path}/creditos.csv")
pagos = pd.read_csv(f"{raw_path}/pagos.csv")
eventos = pd.read_csv(f"{raw_path}/eventos_app.csv")
```

```python
# Limpieza CLIENTES

# Reemplazar ingresos negativos por NaN
clientes['ingreso_mensual_estimado'] = clientes['ingreso_mensual_estimado'].apply(lambda x: np.nan if x < 0 else x)

# Imputar nulos con mediana
clientes['ingreso_mensual_estimado'] = clientes['ingreso_mensual_estimado'].fillna(clientes['ingreso_mensual_estimado'].median())

# Imputar score externo con mediana
clientes['score_externo'] = clientes['score_externo'].fillna(clientes['score_externo'].median())
```

```python
# Limpieza CREDITOS

# Imputar producto_credito nulo con "Desconocido"
creditos['producto_credito'] = creditos['producto_credito'].fillna("Desconocido")
```

```python
# Limpieza PAGOS

# Mantener fecha_pago nula (representa cuotas no pagadas)
# Imputar medio_pago nulo con "No registrado"
pagos['medio_pago'] = pagos['medio_pago'].fillna("No registrado")
```

```python
# Limpieza EVENTOS

# No hay nulos, pero revisar duplicados
eventos = eventos.drop_duplicates()
```

```python
# Guardar datasets procesados

clientes.to_csv(f"{processed_path}/clientes_clean.csv", index=False)
creditos.to_csv(f"{processed_path}/creditos_clean.csv", index=False)
pagos.to_csv(f"{processed_path}/pagos_clean.csv", index=False)
eventos.to_csv(f"{processed_path}/eventos_clean.csv", index=False)
```
